```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.common.by import By

from bs4 import BeautifulSoup as bs
import re
# from tqdm import tqdm

import pandas as pd
import numpy as np
```

```python
# list of target website
urls = ['https://www2.eecs.berkeley.edu/Faculty/Lists/faculty.html' ]
```

```python
def getHTML(url):
    s=Service(ChromeDriverManager().install())
    driver = webdriver.Chrome(service=s)
    driver.get(url)
    content = driver.page_source
    return content

def parseFacultList(html):
    soup= bs(html, 'html.parser')
    fac_list=soup.find_all("div",class_="media")
    nameList=[]
    profile_url=[]
    emailList=[]
    areaList=[]
    for f in fac_list:
        #extract name
        nameList.append(f.find("h3",class_="media-heading").text.strip()) 

        #extract profile ulr
        profile_url.append("https://www2.eecs.berkeley.edu/"+f.find("h3",class_="media-heading").a['href'])

        # extract email
        try:
            email=re.findall(r'[a-zA-Z\.]+@.+.edu',[*f.p.strings][3].strip())[0]
        except:
            email="N/A"
        emailList.append(email)

        # extract area (abbr)
        areaList.append(";".join(re.findall(r'Areas/(\w+)',str(f))))
    df=pd.DataFrame(list(zip(nameList,emailList,areaList,profile_url)),columns=["Name","Email","Area","Profile_url"])
    return df
```

```python
html=getHTML(urls[0])
df=parseFacultList(html)
```

```python
df.to_csv("faculty_contact_list.csv",index=False)
df
```
